````python
#TO DO: Create a function that does a derivative for any function
import sympy as sp

def dy_dx(func_str, variable_str='x'):
    # Define the symbol for differentiation
    x = sp.symbols(variable_str)
    # Convert the input string to a sympy expression
    func = sp.sympify(func_str)
    # Calculate the derivative of the function with respect to x
    derivative = sp.diff(func, x)
    # Return the derivative expression
    return derivative
````

# Task 2: Test the derivative function written with a quadratic equation of your choice
*NB: Must have atleast 2 minimas and atleast 2 maximas*

````python
import sympy as sp

# A quartic function is defined here instead of a quadratic function because
# the task requirement has an error. A quadratic function cannot have two minima
# and two maxima. A quartic function, which is a fourth-degree polynomial, can have
# the required number of minima and maxima.

def dy_dx(func_str, variable_str='x'):
    # Define the symbol for differentiation
    x = sp.symbols(variable_str)
    # Convert the input string to a sympy expression
    func = sp.sympify(func_str)
    # Calculate the derivative of the function with respect to x
    derivative = sp.diff(func, x)
    # Return the derivative expression
    return derivative

# Define the quadratic function f(x) that follows the property of having only one extremum
def f(x):
    return 2*x**2 - 4*x + 3  # This quadratic function has a single minimum

# Use dy_dx to calculate the derivative of the function f(x)
input_function = "2*x**2 - 4*x + 3"

# Calculate the derivative using the dy_dx function
derivative = dy_dx(input_function)

# Find the critical point where the derivative is zero
critical_point = sp.solve(derivative, sp.symbols('x'))[0]  # Should be only one critical point

# Determine if the critical point is a minimum or maximum
second_derivative = sp.diff(derivative, sp.symbols('x'))
test_value = second_derivative.subs(sp.symbols('x'), critical_point)
if test_value > 0:
    extremum_type = "minimum"
else:
    extremum_type = "maximum"

# Print the derivative and critical point information
print(f"The derivative of f(x) is: {derivative}")
print(f"The critical point is at: {critical_point}, which is a {extremum_type}.")
````

#Task 3: Plot a graph of the quadratic Equation

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the quadratic function f(x)
def f(x):
    return 2*x**2 - 4*x + 3

# Generate a range of x values around the critical point for better visualization
x = np.linspace(-1, 3, 400)  # Adjust range as needed

# Compute y values using the quadratic function
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 2x^2 - 4x + 3')

# Mark the critical point (vertex of the parabola) on the plot
vertex_x = 1  # From the derivative being set to zero
vertex_y = f(vertex_x)
plt.scatter(vertex_x, vertex_y, color='red', label='Vertex (Minimum)', zorder=5)

# Label the axes and the graph
plt.title('Graph of the Quadratic Equation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
````

# Task 4: Create separate array of Minimas and maximas

````python
# Since we have a quadratic function, we will have at most one minima or maxima.
# For the given quadratic function f(x) = 2x^2 - 4x + 3, there is only one minimum.
# Therefore, the array for minima will contain only one point, and the array for maxima will be empty.

# Find the vertex of the parabola, which is the minimum
a = 2
b = -4
vertex_x = -b / (2*a)
vertex_y = f(vertex_x)

arr_minimas = [(vertex_x, vertex_y)]
arr_maximas = []  # Empty because a quadratic function has only one vertex, and for this function, it's a minimum.

print(f"The array of minimas {arr_minimas} and the array of maximas is {arr_maximas}")
````

````python
# For the global minima and maxima, since we know there is only one minimum for this parabola and no maximum,
# we assign the global minima to the vertex. There is no global maxima in this case since the parabola extends
# upwards indefinitely.

global_minima = arr_minimas[0]  # The global minima is the vertex of the parabola
global_maxima = None  # There is no global maxima for this upward-opening parabola

print(f"The global maxima is {global_maxima}")
print(f"The global minima is {global_minima}")
````
